```python
#let's transform categorical values into dummies/Convert categorical variable into dummy/indicator variables.

data = pd.get_dummies(data, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
data
```

##### Age - Older patients tend to have higher risk of heart disease. The data shows risk increasing with age.

```python
# Age - Older patients tend to have higher risk of heart disease. The data shows risk increasing with age.

age = data['age']
thalach = data['thalach']

# Create a scatter plot
plt.figure(figsize=(15, 8))
plt.scatter(age, thalach, c='blue', alpha=0.5)
plt.title('Age vs. Maximum Heart Rate (thalach)')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.grid(True)
plt.show()

```

##### Sex - Males have higher risk than females.

##### Chest pain type (cp) - Angina (chest pain) is associated with higher risk.

```python
# CP- whoever feels more chest pain have more risk of heart disease

# Example data (replace this with your actual data)
data = pd.DataFrame({
    'chest_pain_type': ['Typical Angina', 'Atypical Angina', 'Non-Anginal Pain', 'Asymptomatic',
                        'Typical Angina', 'Atypical Angina', 'Non-Anginal Pain', 'Asymptomatic',
                        'Typical Angina', 'Atypical Angina', 'Non-Anginal Pain', 'Asymptomatic'],
    'heart_disease': [1, 0, 1, 0, 1, 0, 0, 1, 1, 1, 0, 1]  # 1 for heart disease, 0 for no heart disease
})

# Create a bar chart to visualize the distribution
plt.figure(figsize=(8, 6))
data.groupby(['chest_pain_type', 'heart_disease']).size().unstack().plot(kind='bar', stacked=True) #treats the remaining variables diffrently 
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.title('Distribution of Chest Pain Types by Heart Disease Status')
plt.xticks(rotation=45)
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()

```

```python
# Create a histogram of cholesterol levels
plt.figure()
plt.hist(data['chol'], bins=20)
plt.xlabel('Cholesterol Level')
plt.ylabel('Count')
plt.title('Cholesterol Level Distribution')

# Show the plots
plt.show()
```
